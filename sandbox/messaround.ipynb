```python
boards = {}
with open('boards.txt', 'r') as f:
    line = f.readline()
    while line != '':
        while line != '' and '.name' not in line:
            lines = f.readlines()

        name = line[line.rfind('=')+1:].rstrip()

        while line != '' and 'vid.' not in line:
            line = f.readline()
            
        while "vid" in line or "pid" in line:
            vid = line[line.rfind('x')+1:].rstrip()
            line = f.readline()
            pid = line[line.rfind('x')+1:].rstrip()
            vid_pid = vid + ":" + pid
            if vid_pid in boards:
                raise RuntimeError('Duplicate board', vid_pid)

            boards[vid_pid] = name
 
            line = f.readline()
```
